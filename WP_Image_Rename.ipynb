```python
#@title Default title text
website_login_url = "https://ephraim.pharrageprojects.co.za/wp-admin" #@param {type:"string"}
login_email_username = "ephraim" #@param {type:"string"}
login_password = "Ephy1992" #@param {type:"string"}

!pip install selenium
!apt-get update 
!apt install chromium-chromedriver

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import re

# You need to download the google chrome driver https://sites.google.com/a/chromium.org/chromedriver/
# Provide a path to the google chrome webdriver, thats the path you downloaded it to
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)


# Fill in your login details to login
# your_website_login_url = input('Enter WP Login URL: ')
# your_email_username = input('Enter Your Username/Email: ')
# your_password = input('Enter Your Password: ')


# The URL your script must visit
#driver.get("https://ephraim.pharrageprojects.co.za/wp-admin")
driver.get(website_login_url)


# Find the username and password fields
email_input = driver.find_element(By.ID, "user_login")
pass_input = driver.find_element(By.ID,"user_pass")
cnt_submit = driver.find_element(By.ID,"wp-submit")

# For the script to be able to login you need to supply your login details
email_input.send_keys(login_email_username)
pass_input.send_keys(login_password)

# Submit login form
cnt_submit.click()

# Wait for page to load everything
driver.implicitly_wait(10)

# Click Media on the menu

try:
  # notNow = driver.find_element(By.LINK_TEXT,'Media')
  notNow = driver.find_element(By.XPATH, "//a[@href ='upload.php']")
  notNow.click()
except NoSuchElementException:
    print("Could not find media on link.")

try:
  # Open the first image
  thumbnail = driver.find_element(By.CLASS_NAME, 'thumbnail')
  thumbnail.click()
except NoSuchElementException:
    print("Could not find thumbanil to click.")

def task_function():

    try:
      # Copy the name of the image form the title input box
      image_text = driver.find_element(By.XPATH,"//*[@data-setting='title']/input[@type='text']").get_attribute('value')
      print("Original image = "+image_text)
      # Remove special characters from the image name
      image_text = re.sub('[^a-zA-Z0-9 \n\.]', ' ', image_text)
      print("Renamed image = "+image_text)
    except NoSuchElementException:
      print("Could not select the image path")

    try:
      # Select the alt input box and paste the image title into it
      alt_text = driver.find_element(By.XPATH,"//*[@data-setting='alt']/input[@type='text']")
      alt_text.clear()
      alt_text.send_keys(image_text.title())
      driver.implicitly_wait(2)
    except NoSuchElementException:
      print("Could not paste title to alt box")

    try:
      # Select the Description input box and paste the image title into it
      desc_text = driver.find_element(By.XPATH,"//*[@data-setting='description']/textarea")
      desc_text.clear()
      desc_text.send_keys(image_text.title())
      driver.implicitly_wait(2)
    except NoSuchElementException:
      print("Could not paste title inside description")

    try:
      # Select the Caption input box and paste the image title into it
      caption_text = driver.find_element(By.XPATH,"//*[@data-setting='caption']/textarea")
      caption_text.clear()
      caption_text.send_keys(image_text.title())
      driver.implicitly_wait(2)
    except NoSuchElementException:
      print("Could not paste title inside Caption")

    try:
      # Select the title input box and paste the cleaned up image title into it
      title_text = driver.find_element(By.XPATH,"//*[@data-setting='title']/input[@type='text']")
      title_text.clear()
      title_text.send_keys(image_text.title())
      driver.implicitly_wait(2)
    except NoSuchElementException:
        print("Could not replace old title")

    try:    
      # Go to next post
      next_post = driver.find_element(By.CLASS_NAME,"right")
      next_post.click()
    except NoSuchElementException:
      print("Could not go to next post")

task_function()

try:
  # Find images and save all of them to a list
  all_posts = driver.find_elements(By.CLASS_NAME,"thumbnail")
 
  for single_post in all_posts:
      # Wait for the post to load
      driver.implicitly_wait(10)
      task_function()
except NoSuchElementException:
  print("Could not loop")

#print(driver.page_source);
```
